```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
x=np.array([5,15,25,35,45,55]).reshape(-1,1)
print(x)
y=np.array([5,20,14,32,22,38])
print(y)
plt.scatter(x,y)
```

```python
model=LinearRegression()
```

```python
model.fit(x,y)
```

```python
print(model.intercept_)
print(model.coef_)
print(model.score(x,y))
```

```python
y_pred=model.predict(x)
print(y_pred)
```

```python
print(model.score(x,y_pred))
plt.scatter(x,y)
```

```python
print(model.intercept_+model.coef_*x)
plt.scatter(x,y)
plt.plot(x,y_pred)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
x=np.array([5,15,25,35,45,55]).reshape(-1,1)
print(x)
y=np.array([5,20,14,32,22,38])
print(y)
plt.scatter(x,y)
model=LinearRegression()
model.fit(x,y)
print(model.intercept_)
print(model.coef_)
print(model.score(x,y))
y_pred=model.predict(x)
print(y_pred)
print(model.score(x,y_pred))
plt.scatter(x,y)
print(model.intercept_+model.coef_*x)
plt.scatter(x,y)
plt.plot(x,y_pred)
```
